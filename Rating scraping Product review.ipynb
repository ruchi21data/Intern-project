```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
#oening web driver and website
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.amazon.in/HP-Pavilion-Graphics-Keyboard-14-dv0054TU/product-reviews/B08WBB369L/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
driver.get(url)
```

```python
#scraping reviews and rating from amazon 
stars = []
reviews_ = []
page_no = 1

while (page_no <45):
    reviews=driver.find_elements_by_xpath("//div[@class='a-section celwidget']")
    for review in reviews:
        try:
            star = review.find_elements_by_xpath(".//a[@class='a-link-normal']")[0].get_attribute('title')
            stars.append(star)
        except:
            stars.append("-")
        try:
            review_text =  review.find_element_by_xpath(".//span[@data-hook='review-body']").text
            reviews_.append(review_text)
        except:
            reviews_.append('-')
    page_no += 1
    try:
        next_page = driver.find_element_by_xpath("//li[@class='a-last']/a").get_attribute('href')
        driver.get(next_page)
    except:
        break
```

```python
len(reviews_)
```

```python
# Make dataframe
df = pd.DataFrame({'Reviews':reviews_,'Ratings':stars})
df
```

```python
#saving as CSV file
df.to_csv('REV1.csv',index=False)
```
